```python
import os
import csv
import pandas as pd
import numpy as np
```

```python
cust_data=pd.read_csv("Cust_data.csv")
```

```python
cust_demo=pd.read_csv("Cust_demo.csv")
```

```python
cust_s1=cust_demo[cust_demo['Martial_Status']=="Married"] [cust_demo['Own_House']==1] [cust_demo['age']>28][['ID', 'age', 'Gender' , 'Location']]
cust_s1.head(5)
```

```python
cust_demo.rename(columns = {'NumberOfDependents':'No_of_dependents'}, inplace = True)
cust_demo.head(5)
```

```python
display(cust_data.drop_duplicates())
display(cust_demo.drop_duplicates())
```

```python
cust_data['No_of_30_plus_DPD']=cust_data['No_of_30_59_DPD']+cust_data['No_of_90_DPD']+ cust_data['No_of_60_89_DPD']
cust_data['No_of_60_plus_DPD'] = cust_data['No_of_90_DPD']+ cust_data['No_of_60_89_DPD']
cust_data.head(5)
```

```python
cust_leftjoin=pd.merge(cust_data, cust_demo, how='left')
cust_leftjoin.head()
```

```python
cust_rightjoin=pd.merge(cust_data, cust_demo, how='right')
cust_rightjoin.head()
```

```python
cust_inner=pd.merge(cust_data, cust_demo, how='inner')
cust_inner.head()
```

```python
cust_outer=pd.merge(cust_data, cust_demo, how='outer')
cust_outer.head()
```

```python
print("no. o"cust_leftjoin['ID'].count())
new_df=cust_inner[cust_inner['Martial_Status']=="Married"]
new_df['ID'].count()
```

```python
print("count of customers:",cust_inner['ID'].count())
print("percentage of customers:",cust_inner['ID'].count()/cust_outer['ID'].count()*100)
print("average of RevolvingUtilization:",cust_inner['RevolvingUtilization'].mean())
print("Standard deviation of RevolvingUtilization:",cust_inner['RevolvingUtilization'].std())
print("average of MonthlyIncome:",cust_inner['MonthlyIncome'].mean())
print("maximum of MonthlyIncome:",cust_inner['MonthlyIncome'].max())
print("Standard deviation of MonthlyIncome:",cust_inner['MonthlyIncome'].std())
print("average of age:",cust_inner['age'].mean())
print("average of No_of_dependents:",cust_inner['No_of_dependents'].mean())
```

```python
new_df=cust_inner[cust_inner['Martial_Status']=="Married"]
print("no. of married customers:",new_df['Martial_Status'].count())
print("percentage of married customers:",new_df['ID'].count()/cust_outer['ID'].count()*100)
```
